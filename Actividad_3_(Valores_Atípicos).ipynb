```python
data=pd.read_csv("Ventas_totales.csv")
data
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data["tarjetas_debito"]=data["tarjetas_debito"].fillna(method="ffill")
data["tarjetas_credito"]=data["tarjetas_credito"].fillna(method="ffill")
data["otros_medios"]=data["otros_medios"].fillna(method="ffill")
data["subtotal_ventas_alimentos_bebidas"]=data["subtotal_ventas_alimentos_bebidas"].fillna(method="ffill")
data["bebidas"]=data["bebidas"].fillna(method="ffill")
data["almacen"]=data["almacen"].fillna(method="ffill")
data["panaderia"]=data["panaderia"].fillna(method="ffill")
data["lacteos"]=data["lacteos"].fillna(method="ffill")
data["carnes"]=data["carnes"].fillna(method="ffill")
data["verduleria_fruteria"]=data["verduleria_fruteria"].fillna(method="ffill")
data["alimentos_preparados_rotiseria"]=data["alimentos_preparados_rotiseria"].fillna(method="ffill")
data["indumentaria_calzado_textiles_hogar"]=data["indumentaria_calzado_textiles_hogar"].fillna(method="ffill")
data["electronicos_articulos_hogar"]=data["electronicos_articulos_hogar"].fillna(method="ffill")
data["otros"]=data["otros"].fillna(method="ffill")
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data.info()
```

```python
cuantitativas=data.iloc[: , 1:23]
cuantitativas
```

```python
cualitativas=data.iloc[ 0 , :]
cualitativas
```

```python
fig = plt.figure(figsize = (15,8))#tama√±o de la figura (largo, ancho)
cuantitativas.plot(kind='box',vert=True)#vertical True, horizontal False
plt.title("valores atipicos del dataframe")#Titulo
plt.show()#lo grafica
```

```python
y = cuantitativas
Limite_Superior = y.mean() + 3*y.std()#establecer limite superior, en este caso 3 desviaciones estandar
Limite_Inferior = y.mean() - 3*y.std()#establecer limite inferior
print("Limite Superior permitido", Limite_Superior)#imprimir limite
print("Limite Inferior permitido", Limite_Inferior)#imprimir limite
```

```python
pd.options.display.max_rows = None
```

```python
Datos_sin_Outliers = cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
Datos_sin_Outliers
```

```python
Valores_nulos = Datos_sin_Outliers.isnull().sum()
Valores_nulos
```

```python
data_clean = Datos_sin_Outliers.copy()
data_clean = data_clean.fillna(round(Datos_sin_Outliers.mean(),1))
data_clean
```

```python
Valores_nulos = data_clean.isnull().sum()
Valores_nulos
```
